```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('superstore.csv')
data.info()
```

```python
total = []

for item in data['Sales']:
    k = str(item)[1:]
    k = k.replace(',', '')
    print(k)
    total.append(int(k, base=10))

data['SalesN'] = pd.Series(total)
data.head()    
```

```python
total = data[['SalesN', 'Region']].groupby('Region').sum()
    
s = pd.DataFrame()
s['Region'] = data['Region'].unique()
s['TotalSales'] = list(total['SalesN'])
```

```python
g = sns.FacetGrid(s)
g.map(plt.bar, 'Region', 'TotalSales')
```

```python
g = sns.FacetGrid(s.sort_values(by='TotalSales'))
g.map(plt.bar, 'Region', 'TotalSales')
```
